```python
from sklearn.model_selection import train_test_split
#from sklearn.cross_validation import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#path = "/content/drive/My Drive/2019_dataset/movement_daily.csv"
#path = "/content/drive/My Drive/2019_dataset/regdata.csv"
data =pd.read_csv("2019_model_dataset.csv")
data
```

```python
vader = data["vader_sent_hourly mean"].values
blob = data["blob_sent_hourly mean"].values
tweetdir = data["tweet dir"].values
tweet_vol = data["Daily Weight count"].values
v_move = data["vsent dir"].values
b_move = data["bsent dir"].values
diff = data['price mov']
multifeat = data.filter(["vader_sent_hourly mean","blob_sent_hourly mean","tweet dir","Daily Weight count"])
```

```python
#Kfold Cross Validation with Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold # import KFold
from sklearn import metrics
features = [vader,blob,tweetdir,tweet_vol,v_move,b_move]

feature = ['vader','blob','tweetdir','tweet_vol','v_move','b_move']
for i in range(len(features)):
    xbar = features[i].reshape(-1,1)
    kf = KFold(n_splits=10) # Define the split - into 2 folds 
    
    kf.get_n_splits(xbar) # returns the number of splitting iterations in the cross-validator
    #KFold(n_splits=10, random_state=None, shuffle=False)
    scores = []
    mat = np.zeros([2,2])
    for train_index, test_index in kf.split(xbar):
        #print("TRAIN:", train_index, "TEST:", test_index)
        X_train, X_test = xbar[train_index], xbar[test_index]
        y_train, y_test = diff[train_index], diff[test_index]
        logisticRegr = LogisticRegression()
        logisticRegr.fit(X_train, y_train)

        predictions = logisticRegr.predict(X_test)
        score = logisticRegr.score(X_test, y_test)
        scores.append(score)
        #print("{} Score : {}".format(feature[i],score))
        cm = metrics.confusion_matrix(y_test, predictions)
        mat += cm
        #print(cm)
    idx = np.mean(scores)
    print("{} Score : {}".format(feature[i],idx))

    print(mat)

```

```python
#multi feature logistic regression
x_train, x_test, y_train, y_test = train_test_split(multifeat, diff, test_size=0.20, random_state=0)
logisticRegr = LogisticRegression()
logisticRegr.fit(x_train, y_train)
predictions = logisticRegr.predict(x_test)
score = logisticRegr.score(x_test, y_test)
print("Multi feat Score : {}".format(score))
cm = metrics.confusion_matrix(y_test, predictions)
print(cm)
```

```python
#svm useless
from sklearn.svm import SVC
#â€˜linearâ€™, â€˜polyâ€™, â€˜rbfâ€™, â€˜sigmoidâ€™
kernels = ['linear','rbf','sigmoid']
for i in range(len(features)):
    sbar = features[i].reshape(-1,1)
    x_train, x_test, y_train, y_test = train_test_split(sbar, data["price mov"], test_size=0.20, random_state=0)
    for k in kernels:
        for deg in range(6):
            clf = SVC(kernel = k,gamma ="auto",degree = deg)
            clf.fit(x_train,y_train)
            predictions = clf.predict(x_test)
            score = clf.score(x_test,y_test)
            print("feature used {} with kernel {} with degree {} has score {}".format(feature[i],k,deg,score))
            cm = metrics.confusion_matrix(y_test,predictions)
            print(cm)
```

```python
from sklearn import neighbors

for i in range(len(features)):
    for j in range(1,6):
        for weights in ["uniform","distance"]:
            print(weights,j)
            print("Working on {}".format(feature[i]))
            sbar = features[i].reshape(-1,1)
            x_train, x_test, y_train, y_test = train_test_split(sbar, data["price mov"], test_size=0.20, random_state=0)
            knn = neighbors.KNeighborsClassifier(j, weights=weights)
            knn.fit(x_train, y_train)
            predictions = knn.predict(x_test)
            score = knn.score(x_test,y_test)
            print("The score is : {}".format(score))
            cm = metrics.confusion_matrix(y_test,predictions)
            print(cm)
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

features = [vader,blob,tweetdir,tweet_vol,v_move,b_move]

#X, y = load_iris(return_X_y=True)
feature = ['vader','blob','tweetdir','tweet_vol','v_move','b_move']
for i in range(len(features)):
    scores = []
    mat = np.zeros([2,2])
    for j in range(1,10):
        print("Working on {} {} estimators".format(feature[i],j*100))
        sbar = features[i].reshape(-1,1)
        x_train, x_test, y_train, y_test = train_test_split(sbar, data["price mov"], test_size=0.20, random_state=0)
        regressor = RandomForestClassifier(n_estimators=j*10, random_state=0,max_depth=2)
        regressor.fit(x_train, y_train)
        y_pred = regressor.predict(x_test)
        
        #predictions = knn.predict(x_test)
        score = regressor.score(x_test,y_test)
        #scores.append(score)
        print(confusion_matrix(y_test,y_pred))
        
        print(classification_report(y_test,y_pred))
        print(accuracy_score(y_test, y_pred))
        #print(scores)
        #print(mat)
    #print("The score is : {}".format(score))
    #cm = metrics.confusion_matrix(y_test,predictions)
    #print(cm)
```

```python
#multifeat random forest
x= multifeat
y = data["price mov"]
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
for i in range(1,10):
    classifier = RandomForestClassifier(n_estimators=i*100,max_depth=2, random_state=0)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    print(confusion_matrix(y_test,y_pred))
    print(classification_report(y_test,y_pred))
    print(accuracy_score(y_test, y_pred))
```
