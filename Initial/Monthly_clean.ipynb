```python
import pandas as pd
import re
```

```python
df = pd.read_csv("tweet.csv")
df.head(10)
data = df.drop(["mentions","to","replies","geo","permalink","id","hashtags"],axis=1)
data["date2"] = pd.to_datetime(data["date"],errors= "coerce").dt.date
data["month"] = pd.to_datetime(data["date"],errors= "coerce").dt.month
df.head(10)
data_group = data.groupby(data['date2']).agg("count")
print(data_group.head(30))
print(data)
```

```python
def clean(data):
    #cleanr = re.compile("<.*?>")
    cleanr= re.sub(r"http\S+", "", data)
    #cleanr= re.sub(cleanr, "", data)
    #cleantext = re.sub(cleanr,'',data)
    cleantext = re.sub('[^a-zA-Z# ]','',cleanr)
    clean_hash = re.sub(r"#\S+","",cleantext)
    
    return clean_hash
for i in range(data["text"].shape[0]):
#for i in range(20):
    try:
        data["text"][i]= clean(data["text"][i])
        if(i%100==0):
            print("this is the {} th tweet ".format(i))
    except:
        print("This tweet got problem {}".format(data["text"][i]))
        pass
```
